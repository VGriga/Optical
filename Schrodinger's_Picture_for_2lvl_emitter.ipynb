```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
%matplotlib inline
```

```python
#non time dependent parameters of system, including O-Raby, difference of Energy between levels d=E2-E1

O=1
d=1

#initial values for elements of density matrix

r0= np.array([1, 0 ,0, 0], dtype = "complex_")
#system of equations
def systemEq(d,v): 
    return [-1j*O*(v[1]-v[2]),           #r11
            1j*d*v[1]+1j*O*(v[3]-v[0]),  #r12
            -1j*d*v[2]+1j*O*(v[0]-v[3]), #r21
            1j*O*(v[1]-v[2])]            #r22

#solver (function,limits of integration, initial values, range of points to calculate)

res = solve_ivp(fun=systemEq, t_span=[0, 5], y0=r0 ,t_eval=np.linspace(0,5,200))
#extracting time

t=res.t

#extracting solution of system 
r11=res.y[0]
r12=res.y[1]
r21=res.y[2]
r22=res.y[3]

#calculating Bloch parameters
N=r11-r22
R=r12+r21
P=-1j*(r12-r21)

fig = plt.figure(figsize=(6,6))
plt.plot(t,r11.real,label='r11,Re')
plt.plot(t,r12.real,label='r12,Re')
plt.plot(t,r21.real,label='r21,Re')
plt.plot(t,r22.real,label='r22,Re')
plt.plot(t,r11.real+r22.real,label='r11+r22')
plt.grid()
plt.xlabel('t')
plt.ylabel('r')
plt.title('Density matrix elements, Real')
plt.legend()
plt.show()


fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(r11.real, r11.imag, t,label='r11')
ax.scatter(r12.real, r12.imag, t,label='r12')
ax.scatter(r21.real, r21.imag, t,label='r21')
ax.scatter(r22.real, r22.imag, t,label='r21')
ax.set_xlabel('Re')
ax.set_ylabel('Im')
ax.set_zlabel('t')
ax.legend()
plt.title('Density matrix elements')
plt.show()


plt.plot(t,R.real,label='R')
plt.plot(t,N.real,label='N')
plt.plot(t,P.real,label='P')
plt.grid()
plt.xlabel('t')
plt.ylabel('N,R,P')
plt.title('Coefficients, time dependence')
plt.legend(loc='lower left')
plt.show()



fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(N.real, R.imag, P.real)
ax.set_xlabel('N')
ax.set_ylabel('R')
ax.set_zlabel('P')
plt.title('Bloch system')
plt.show()




```
